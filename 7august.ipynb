```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

# Task 1: Create a Distribution Plot

```python
data=np.random.normal(size=300)
sns.histplot(data, kde=True, bins=30,color="orange")

# Show the plot
plt.title("Distribution plot")
plt.xlabel("Value")
plt.ylabel("frequence")
plt.show()
```

# Task 2: Create a Joint Plot

```python
data=pd.DataFrame({
'x':np.random.normal(size=100),
'y':np.random.normal(size=100)
})
    
sns.jointplot(x='x',y='y', kind='scatter', data=data,color="orange")
plt.tittle('joint plot')
plt.show()
```

# Task 3: Create a Boxplot

```python
data=pd.DataFrame({
    'category':['a','b','c']*20,
    'value':np.random.rand(60)
})
sns.boxplot(x='category',y='value',data=data)
plt.show()
```

# Task 4: Create a Violin Plot

```python
data=pd.DataFrame({
    'category':['a','b','c']*20,
    'value':np.random.rand(60)
})
sns.catplot(x='category',y='value',data=data ,kind='violin',camp=')
plt.show()
```

# Task 5: Create a Countplot

```python
data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
sns.countplot(data)
plt.show()
```

# Task 6: Create a Heatmap

```python
data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
sns.heatmap(data, annot=True, cmap='viridis')
plt.show()
```

# Task 7: Create a Rug Plot

```python
data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
sns.rugplot(data)
plt.show()
```

# Task 8: Create a Swarm Plot

```python
data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
sns.swarmplot(data)
plt.show()
```

# Task 9: Create a Factor Plot

```python
data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
sns.catplot(data,  cmap='viridis')
plt.show()
```
