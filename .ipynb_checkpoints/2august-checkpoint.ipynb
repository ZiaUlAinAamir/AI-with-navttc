```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 6, 8, 10]
plt.plot(x1, y1, color='blue', label='Line 1')
plt.plot(x2, y2, color='green', label='Line 2')
# Add titles and labels
plt.title('Line Plot with Different Colors')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()









```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 6, 8, 10]
y3 =  [1,3,4,6,6] 
plt.plot(x1, y1, color='blue', label='Line 1',linestyle="--",marker="o" ,)
plt.plot(x2, y2, color='green', label='Line 2',linestyle="-.",marker="o" )
plt.plot(x1, y3, color='red', label='Line 2',linestyle="solid",marker="o" )
plt.title('Line Plot with Different Colors')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()


```

# task 7

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
x1=[2,3,5,6,7]
y2=[2,4,6,7,8]
x3=[2,3,4,6,7]
y3=[8,6,6,5,4]
plt.plot(x, y, linewidth=2, marker='o', markersize=10, color='b', linestyle='-', label='Data Line')
plt.plot(x1, y1, linewidth=2, marker='^', markersize=10, color='green', linestyle='-.', label='Data Line')
plt.plot(x2, y2, linewidth=2, marker='s', markersize=10, color='yellow', linestyle='--', label='Data Line')
plt.plot(x3, y3, linewidth=2, marker='^', markersize=10, color='red', linestyle=':', label='Data Line')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Sample Plot Title')
plt.show()
```

```python
x = [1, 2, 3, 4, 5,2 ,3,1,4,3,5,5,5,6,6,6,5,7,6,8,8,9,9,9,9]
y = [2,3,5,6,7,8,8,9,8,7,6,5,4,3,3,3,2,2,1,2,3, 3, 5, 7, 1,1]
plt.hist(x, bins =9 , linewidth=2, color='navy', label='Data Line', alpha = 0.5)
plt.hist(y, bins =9 , linewidth=2, color='green', label='Data Line', alpha = 0.5)

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Sample Plot Title')
plt.legend() 

plt.show()
```
