```python
 ##modules we'll use
import pandas as pd
import numpy as np

# for Box-Cox Transformation
from scipy import stats

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# height and width of figure
plt.figure(figsize=(10,5))

# Line chart showing
sns.lineplot(df)
```

```python
##Load the tips dataset
tips = sns.load_dataset("tips")

# Create a scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Show the plot
plt.title("Total Bill vs. Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
 ##Create a box plot
sns.boxplot(x="day", y="total_bill", data=tips)

# Show the plot
plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
data=np.random.normal(size=300)
sns.histplot(data, kde=True, bins=30,color="green")

# Show the plot
plt.title("Distribution plot")
plt.xlabel("Value")
plt.ylabel("frequence")
plt.show()
```

```python
data=pd.DataFrame({
'x':np.random.normal(size=100),
'y':np.random.normal(size=100)
})
    
sns.jointplot(x='x',y='y', kind='scatter', data=data,color="green")
plt.tittle('joint plot')
plt.show()
```

```python
sns.jointplot(x='x',y='y', kind='hist', data=data,color="green")
```

```python
sns.jointplot(x='x',y='y', kind='hex', data=data,color="green")
```

```python
sns.jointplot(x='x',y='y', kind='reg', data=data,color="green")
```

```python
sns.jointplot(x='x',y='y', kind='kde', data=data,color="green")
```
